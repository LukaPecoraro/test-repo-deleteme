```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument("--headless")

#driver = webdriver.Chrome(service=service, options=options)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)


URL = "https://fri.uni-lj.si/sl/o-fakulteti/osebje"
driver.get(URL)


elements = driver.find_elements(By.CLASS_NAME, "listedProf")
prof_page_links = []

for element in elements:
    # nested, links inside the <ul> elements
    links = element.find_elements(By.TAG_NAME, "a")
    print(element.text)
    for link in links:
        url = link.get_attribute("href")
        print(url)
        prof_page_links.append(url)


driver.quit()
```

#### Beautiful soup approach
- Get all lists of staff starting with letter
- Find links in pages corresponding to staff
- Iterate through the links, get every page html with request
- Parse the site for info

```python
import requests
from bs4 import BeautifulSoup



URL = "https://fri.uni-lj.si/sl/o-fakulteti/osebje"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
#results = soup.find(id="ResultsContainer")
# print(results.prettify())


staff_letter_list = soup.find_all("ul", class_ = "listedProf")
print(staff_letter_list[0].prettify())
```

```python
staff_letter_list[0].find_all("a")
```

```python
# find all links for staff pages ion the letter list

staff_page_urls = [] 

for staff_letter in staff_letter_list:
    # FIND ALL LINKS WITH A HREF
    urls = staff_letter.find_all("a", href=True)
    urls = [el["href"] for el in urls] # extract only the href url from the link
    
    staff_page_urls.extend(urls) # adds all elements in a list
    # for url in urls:
    #     staff_page_urls.append(url)

    ##### ALTERNATIVE WAY #############
    #if link.has_attr('href'):
    #    print(link['href'])

print(staff_page_urls)
```

```python
names = []
phones = []

for person_url in staff_page_urls:
    page = requests.get(f"https://fri.uni-lj.si{person_url}")
    soup = BeautifulSoup(page.content, "html.parser")
    person_name = soup.find(id="profName")
    print(person_name.text.strip())
    names.append(person_name.text.strip())
    phone = soup.find(id="telephoneInfo")
    if phone is not None:
        print(phone.text.strip()[3:])
        phones.append(phone.text.strip()[3:])
    else:
        phones.append("")
    
    #print(phone.text.strip()[3:])



data_dict = {"name" : names, "phone": phones}

pd.DataFrame(data_dict)
```
